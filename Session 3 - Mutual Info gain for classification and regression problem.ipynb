```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import load_boston
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_wine
```

```python
data=load_wine()
df=pd.DataFrame(data.data,columns=data.feature_names)
df['Target']=data.target
df
```

```python
df['Target'].value_counts()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
X=df.drop(columns=['Target'],axis=1)
y=df[['Target']]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=1)
```

### Mutual Info Gain for Classification Problem 

```python
from sklearn.feature_selection import mutual_info_classif
# Determine the mutual information 
mutual_info=mutual_info_classif(X_train,y_train)
mutual_info
```

```python
mutual_info=pd.Series(mutual_info)
mutual_info.index=X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
from sklearn.feature_selection import SelectKBest

sel_five_cols=SelectKBest(mutual_info_classif,k=5)
sel_five_cols.fit(X_train,y_train)
X_train.columns[sel_five_cols.get_support()]
```

```python
house_df=pd.read_csv('C:\\Users\\Dell\\Data Science\\CampusX Tutorial\\Feature Selection\\test.csv')
house_df_feature=pd.read_csv('C:\\Users\\Dell\\Data Science\\CampusX Tutorial\\Feature Selection\\sample_submission.csv')

house_df.head()
```

```python
house_df_int= [columns for columns in house_df 
               if columns not in house_df.columns[house_df.dtypes=='object']]
len(house_df_int)
```

```python
house_df=house_df[house_df_int]
house_df.drop(columns=['Id'],inplace=True)
house_df
```

```python
house_df['SalesPrice']=house_df_feature['SalePrice']
```

```python
X=house_df.drop(columns=['SalesPrice'],axis=1)
y=house_df[['SalesPrice']]
```

```python
X.fillna(0,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.feature_selection import mutual_info_regression

mutual_info= mutual_info_regression(X_train.fillna(0),y_train)
```

```python
mutual_info=pd.Series(mutual_info)
mutual_info.index=X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
from sklearn.feature_selection import SelectPercentile
select_top_col = SelectPercentile(mutual_info_regression,percentile=20)
select_top_col.fit(X_train,y_train)
X_train.columns[select_top_col.get_support()]
```
